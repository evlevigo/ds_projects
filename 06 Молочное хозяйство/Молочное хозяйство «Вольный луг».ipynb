```python
def nice_t(data):
    
    #Функция для форматирования вывода таблиц
    th_props = [
      ('font-size', '11px'),
      ('text-align', 'center'),
      ('font-weight', 'bold'),
      ('color', 'black'),
      ('background-color', '#40E0D0')
      ]
    td_props = [
      ('font-size', '12px'),
      ('text-align', 'center')
      ]
    tr_props = [
      ('background-color', '#40E0D0'),
      ('text-align', 'center')  
        ]
    tr_nth_child_props = [
      ('background-color', '#CFECEC')  
        ]

    styles = [
      dict(selector="th", props=th_props),
      dict(selector="td", props=td_props),
      dict(selector="tr:nth-child(even)", props=tr_nth_child_props),
      dict(selector="tr:hover", props=tr_props)  
      ]
    display(data.style.set_table_styles(styles))
```

### Функция фильтрации выбросов числовых данных

[К оглавлению](#Оглавление)

```python
def remove_outlier(data, exclude_col):
    
    #Функция исключает строки со значениями, превышающими заданные
    #в процессе корректируется значение нижнего и верхнего порога, чтобы сохранить как можно больше данных
    
    for i in data.select_dtypes(include={'int64', 'float64'}):
        if i not in exclude_col:
            q1 = data[i].quantile(0.20) 
            q3 = data[i].quantile(0.75) 
            iqr = q3-q1 #межквартильный размах
            fence_low  = q1-1.5*iqr #нижняя граница
            fence_high = q3+1.5*iqr #верхняя граница
            data.loc[:, i] = data.loc[(data[i] > fence_low) & (data[i] < fence_high)] #фильтрация по нижней и верхней границе
    return data
```

```python
def data_numbers_analyze(data, exclude_col, x_labels, y_labels):
    
    '''
    Функция выводит общую информацию о датафрейме(describe)
    Выводит диаграммы размаха с группировкой по категориям

    '''
    #создадим датафрейм без столбцов exclude_col
    data_excluded = data.drop(exclude_col, axis=1)
    #создадим датафрейм с наименованиями labels
    df_labels = pd.DataFrame(columns=data.columns.to_list())
    df_labels.loc[len(df_labels)] = x_labels
    df_labels.loc[len(df_labels)] = y_labels
    #создадим цикл для вывода информации
    print ('\nИсследование числовых значений\n')
    print('\nОбщая информация о данных (describe)\n')
    nice_t(data_excluded.describe())
    #создадим переменные для цикла
    x = 0
    y = 0
    #задаём количество строк и столбцов в subplots
    cols_number = 3
    rows_number = ceil(len(data_excluded.select_dtypes(include={'int64', 'float64'}).columns)/cols_number)
    #зададим параметры для вывода диаграмм
    fig, axes = plt.subplots(nrows=rows_number, ncols=cols_number, figsize=(15,15))
    plt.subplots_adjust(wspace=0.25, hspace=0.3)
    #создадим цикл для вывода диаграмм размаха
    for i in data_excluded.select_dtypes(include={'int64', 'float64'}): 
        if y == cols_number:
            x+=1   
            y=0
        sns.set_style("darkgrid")
        boxplot = sns.boxplot(y=data_excluded[i], ax=axes[x,y], width=0.3)
        boxplot.axes.set_title(f"Диаграмма размаха {df_labels[i].values[0]}", fontsize=10)
        boxplot.set_xlabel(df_labels[i].values[0], fontsize=10)
        boxplot.set_ylabel(df_labels[i].values[1], fontsize=10)
        y+=1
```

### Функция вывода гистограммы распределения числовых значений

[К оглавлению](#Оглавление)

```python
def data_hist_analyze(data, exclude_col, x_labels, y_labels):
  
    '''
    Функция выводит гистограммы
    Для статистической проверки принадлежности нормальному распределению применим тест Шапиро-Уилка. 
    Нулевая гипотеза H0 теста Шапиро–Уилка заключается в том, что случайная величина, 
    выборка x которой известна, распределена по нормальному закону. Альтернативная гипотеза H1 заключается в том,
    что закон распределения не является нормальным. 
    Величина близкая к 1 означает, что распределение близко к нормальному
    '''    
    #создадим датафрейм без столбцов exclude_col
    data_excluded = data.drop(exclude_col, axis=1)
    #создадим датафрейм с наименованиями labels
    df_labels = pd.DataFrame(columns=data.columns.to_list())
    df_labels.loc[len(df_labels)] = x_labels
    df_labels.loc[len(df_labels)] = y_labels

    #создадим цикл для оценки распределения числовых данных при помощи histplot
          
    for j in data_excluded.select_dtypes(include={'int64', 'float64'}):
        fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8,5))
        sns.set_style("darkgrid")
        histplot = sns.histplot(data=data_excluded, x=j, color="blue", label=j, kde='true', bins='auto')
        histplot.axes.set_title(f'Гистограмма распределения {df_labels[j].values[0]}', fontsize=10)
        histplot.set_xlabel(df_labels[j].values[0], fontsize=10)
        histplot.set_ylabel('Кол-во', fontsize=10)
        plt.show()
        res = st.shapiro(data_excluded[j])
        print('Тест Шапиро-Уилка',round(res.statistic,4))
```

```python
def data_corr_analyze(data, target_attribute, exclude_col, x_labels, y_labels):
    
    #функция строит графики корреляции
    
    #создадим датафрейм без столбцов exclude_col
    data_excluded = data.drop(exclude_col, axis=1)
    #создадим датафрейм с наименованиями labels
    df_labels = pd.DataFrame(columns=data.columns.to_list())
    df_labels.loc[len(df_labels)] = x_labels
    df_labels.loc[len(df_labels)] = y_labels
    #задание количество столбцов
    ncols = 3
    # расчет количества строк
    nrows = len(data_excluded.select_dtypes(include={'int64', 'float64'}).columns) // ncols + \
           (len(data_excluded.select_dtypes(include={'int64', 'float64'}).columns) % ncols > 0)
    

    for category in data.select_dtypes(exclude={'int64', 'float64'}):
        plt.figure(figsize=(15, 10))
        plt.suptitle(f"Группировка по {df_labels[category].values[0]}", fontsize=18, y=1)
        for i, column in enumerate(data_excluded.select_dtypes(include={'int64', 'float64'}).drop(target_attribute, axis=1)):
            ax = plt.subplot(nrows, ncols, i + 1)
            scatplot = sns.scatterplot(data=data_excluded, 
                               x=target_attribute, 
                               y=column, 
                               hue=category,
                               palette="Paired_r",
                               alpha = 0.6,
                               ax  = ax)
            scatplot.set_xlabel('Удой', fontsize=10)
            scatplot.set_ylabel(df_labels[column].values[0], fontsize=10)
        #plt.tight_layout()    
        plt.show()
```

Файл [`ferma_main.csv`] содержит данные о стаде фермера на текущий момент. Описание данных:
    
   -   `id` — уникальный идентификатор коровы.
   -   `Удой, кг` — масса молока, которую корова даёт в год (в килограммах).
   -   `ЭКЕ (Энергетическая кормовая единица)` — показатель питательности корма коровы.
   -   `Сырой протеин, г` — содержание сырого протеина в корме (в граммах).
   -   `СПО (Сахаро-протеиновое соотношение)` — отношение сахара к протеину в корме коровы.
   -   `Порода` — порода коровы.
   -   `Тип пастбища` — ландшафт лугов, на которых паслась корова.
   -   `порода папы_быка` — порода папы коровы.
   -   `Жирность,%` — содержание жиров в молоке (в процентах).
   -   `Белок,%` — содержание белков в молоке (в процентах).
   -   `Вкус молока` — оценка вкуса по личным критериям фермера, бинарный признак (`вкусно`, `не вкусно`).
   -   `Возраст` — возраст коровы, бинарный признак (`менее_2_лет`, `более_2_лет`).
    
Признаки в датасете `ferma_main.csv` можно разделить на группы.
    
   -   Характеристики коровы: `id`, `Порода`, `порода папы_быка`, `Возраст`.
   -   Характеристики корма: `ЭКЕ (Энергетическая кормовая единица)`, `Сырой протеин, г`, `СПО (Сахаро-протеиновое соотношение)`.
   -   Характеристика пастбища: `Тип пастбища`.
   -   Характеристики молока: `Удой, кг`, `Жирность,%`, `Белок,%`, `Вкус молока`.

### Описание данных cow_buy.csv

Файл [`cow_buy.csv`] — это данные о коровах «ЭкоФермы», которых фермер хочет изучить перед покупкой. Описание данных:
    
   -   `Порода` — порода коровы.
   -   `Тип пастбища` — ландшафт лугов, на которых паслась корова.
   -   `порода папы_быка` — порода папы коровы.
   -   `Имя_папы` — имя папы коровы.
   -   `Текущая_жирность,%` — содержание жиров в молоке (в процентах).
   -   `Текущий_уровень_белок,%` — содержание белков в молоке (в процентах).
   -   `Возраст` — возраст коровы, бинарный признак (`менее_2_лет`, `более_2_лет`).
    
Данные в файлах `ferma_main.csv` и `cow_buy.csv` похожи, но у датасета `cow_buy.csv` несколько особенностей.
    
   -   Содержание белков и жиров в молоке указано на момент продажи — когда «ЭкоФерма» давала коровам свой корм.
   -   Параметры корма `ЭКЕ (Энергетическая кормовая единица)`, `Сырой протеин, г` и `СПО (Сахаро-протеиновое соотношение)` отсутствуют. Технологи заказчика пересмотрели подход к кормлению: для новых коров планируется увеличить значения каждого из этих параметров на 5%.
   -   Кроме того, отсутствуют признаки `Удой, кг` и `Вкус молока`. Это целевые признаки, именно их предстоит предсказать!

#### Разбейте данные на тренировочную и тестовую выборки.

Подготовим датафрейм для дальнейшей работы.

```python

#Сохранение признаков
X = df_ferma_main.drop(['milk_yield','id', 'fat_content_ratio', 'protein_content_ratio', 'milk_taste'], axis=1)
y = df_ferma_main['milk_yield']

# разделение на тренировочную и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y, 
    random_state=RANDOM_STATE
)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
nice_t(X.head())
```

```python
num_col_names = X.select_dtypes(include={'int64', 'float64'}).columns.to_list()
cat_col_names = X.select_dtypes(exclude={'int64', 'float64'}).columns.to_list()
```

<span style='color:red'>---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</span>

#### Кодируйте категориальные признаки с помощью OneHotEncoder.

```python
# создадим переменную с экземпляром класса OneHotEncoder()
# включим параметр для избегания появления дамми-ошибок,
# отключим sparse-отображение
encoder = OneHotEncoder(drop='first', sparse=False) 

# обучим и трансформируем категориальные признаки из тренировочной выборки 
X_train_ohe = encoder.fit_transform(X_train[cat_col_names]) 

# выполним преобразование категориальных признаков в тестовой выборке
X_test_ohe = encoder.transform(X_test[cat_col_names]) 

# получим от кодировщика список новых столбцов и сохраним в новой переменной 
encoder_col_names = encoder.get_feature_names() 
```

<span style='color:red'>---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</span>

#### Масштабируйте количественные признаки с помощью StandardScaler.

```python
# создадим переменную с экземпляром класса StandartScaler()
scaler = StandardScaler()

# обучим и трансформируем количественные признаки из тренировочной выборки 
X_train_scaled = scaler.fit_transform(X_train[num_col_names])

# выполним преобразование количественных признаков в тестовой выборке
X_test_scaled = scaler.transform(X_test[num_col_names])
```

<span style='color:red'>---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</span>

#### Обучите модель линейной регрессии на подготовленных данных.

```python
# создаём тренировочный и тестовый датафреймы из закодированных и отмасштабированных данных
X_train_ohe = pd.DataFrame(X_train_ohe, columns=encoder_col_names)
X_test_ohe = pd.DataFrame(X_test_ohe, columns=encoder_col_names)

X_train_scaled = pd.DataFrame(X_train_scaled, columns=num_col_names)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=num_col_names)

# объединяем все преобразованные данные в тренировочный и тестовый датафреймы
# сбрасываем индексы в датафрейме с числовыми индексами, чтобы избежать ошибок

X_train = pd.concat([X_train_ohe, X_train_scaled], axis=1)
X_test = pd.concat([X_test_ohe, X_test_scaled], axis=1)
```

```python
# выведем пять первых строк из тренировочного датафрейма
nice_t(X_train.head())
```

```python
# инициализация и обучение модели
model_lr_1  = LinearRegression()
model_lr_1.fit(X_train, y_train)

# предсказанные значения для X_test
predictions_1 = model_lr_1.predict(X_test)

# вывод первых предсказанного и настоящего значений
print(predictions_1[0])
print(y_test.reset_index(drop=True)[0])
```

<span style='color:red'>---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</span>

#### Оцените качество модели на тестовой выборке: рассчитайте R², проанализируйте остатки с помощью графиков.

```python
# рассчитаем значение R2 для линейной регрессии
r2 = r2_score(y_test, predictions_1)  
print('- R^2:', round(r2,4))

# получение значения MSE
mse = mean_squared_error(y_test, predictions_1) 
print('- MSE:', round(mse,4))

# получение значения RMSE
rmse = mean_squared_error(y_test, predictions_1, squared=False)
print('- RMSE:', round(rmse,4))

# рассчитайте MAE
mae = mean_absolute_error(y_test, predictions_1) 
print('- MAE:', round(mae,4))
```

```python
# рассчитаем остатки
residuals_1 = y_test - predictions_1

# построим графики по количественным признакам — гистограмму частотности 
# распределения и диаграмму рассеяния
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15,8))
axes[0].hist(residuals_1, bins = 20)
axes[0].set_title('Гистограмма распределения остатков')
axes[0].set_xlabel('Остатки')

axes[1].scatter(x=predictions_1, y=residuals_1)
axes[1].set_xlabel('Предсказания модели')
axes[1].set_ylabel('Остатки')
axes[1].set_title('Анализ дисперсии')
plt.show()
```

```python
#Сохранение признаков
X = df_second_model.drop(['milk_yield','id', 'fat_content_ratio', 'protein_content_ratio', 'milk_taste'], axis=1)
y = df_second_model['milk_yield']

# разделение на тренировочную и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y, 
    random_state=RANDOM_STATE
)

num_col_names = X.select_dtypes(include={'int64', 'float64'}).columns.to_list()
cat_col_names = X.select_dtypes(exclude={'int64', 'float64'}).columns.to_list()

# создадим переменную с экземпляром класса OneHotEncoder()
# включим параметр для избегания появления дамми-ошибок,
# отключим sparse-отображение
encoder = OneHotEncoder(drop='first', sparse=False) 

# обучим и трансформируем категориальные признаки из тренировочной выборки 
X_train_ohe = encoder.fit_transform(X_train[cat_col_names]) 

# выполним преобразование категориальных признаков в тестовой выборке
X_test_ohe = encoder.transform(X_test[cat_col_names]) 

# получим от кодировщика список новых столбцов и сохраним в новой переменной 
encoder_col_names = encoder.get_feature_names() 

# создадим переменную с экземпляром класса StandartScaler()
scaler = StandardScaler()

# обучим и трансформируем количественные признаки из тренировочной выборки 
X_train_scaled = scaler.fit_transform(X_train[num_col_names])

# выполним преобразование количественных признаков в тестовой выборке
X_test_scaled = scaler.transform(X_test[num_col_names])

# создаём тренировочный и тестовый датафреймы из закодированных и отмасштабированных данных
X_train_ohe = pd.DataFrame(X_train_ohe, columns=encoder_col_names)
X_test_ohe = pd.DataFrame(X_test_ohe, columns=encoder_col_names)

X_train_scaled = pd.DataFrame(X_train_scaled, columns=num_col_names)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=num_col_names)

# объединяем все преобразованные данные в тренировочный и тестовый датафреймы
# сбрасываем индексы в датафрейме с числовыми индексами, чтобы избежать ошибок

X_train = pd.concat([X_train_ohe, X_train_scaled], axis=1)
X_test = pd.concat([X_test_ohe, X_test_scaled], axis=1)

# инициализация и обучение модели
model_lr_2  = LinearRegression()
model_lr_2.fit(X_train, y_train)

# предсказанные значения для X_test
predictions_2 = model_lr_2.predict(X_test)

# вывод первых предсказанного и настоящего значений
print('Предсказанное значение:', round(predictions_2[0],2))
print('Значение из тестовой выборки: ', y_test.reset_index(drop=True)[0])

# рассчитаем значение R2 для линейной регрессии
r2 = r2_score(y_test, predictions_2)  
print('- R^2:', round(r2,4))

# получение значения MSE
mse = mean_squared_error(y_test, predictions_2) 
print('- MSE:', round(mse,4))

# получение значения RMSE
rmse = mean_squared_error(y_test, predictions_2, squared=False)
print('- RMSE:', round(rmse,4))

# рассчитайте MAE
mae = mean_absolute_error(y_test, predictions_2) 
print('- MAE:', round(mae,4))

# рассчитаем остатки
residuals_2 = y_test - predictions_2

# построим графики по количественным признакам — гистограмму частотности 
# распределения и диаграмму рассеяния
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15,8))
axes[0].hist(residuals_2, bins = 20)
axes[0].set_title('Гистограмма распределения остатков')
axes[0].set_xlabel('Остатки')

axes[1].scatter(x=predictions_2, y=residuals_2)
axes[1].set_xlabel('Предсказания модели')
axes[1].set_ylabel('Остатки')
axes[1].set_title('Анализ дисперсии')
plt.show()
```

```python
#Сохранение признаков
X = df_log_model.drop(['milk_yield','id', 'milk_taste'], axis=1)
y = df_log_model['milk_taste']

# разбиваем данные на тренировочную и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y, 
    random_state=RANDOM_STATE
)

# создадим переменную с экземпляром класса OneHotEncoder()
# включим параметр для избегания появления дамми-ошибок,
# отключим sparse-отображение
encoder = OneHotEncoder(drop='first', sparse=False) 

# обучим и трансформируем категориальные признаки из тренировочной выборки 
X_train_ohe = encoder.fit_transform(X_train[cat_col_names]) 

# выполним преобразование категориальных признаков в тестовой выборке
X_test_ohe = encoder.transform(X_test[cat_col_names]) 

# получим от кодировщика список новых столбцов и сохраним в новой переменной 
encoder_col_names = encoder.get_feature_names() 

# создадим переменную с экземпляром класса StandartScaler()
scaler = StandardScaler()

# обучим и трансформируем количественные признаки из тренировочной выборки 
X_train_scaled = scaler.fit_transform(X_train[num_col_names])

# выполним преобразование количественных признаков в тестовой выборке
X_test_scaled = scaler.transform(X_test[num_col_names])

# создаём тренировочный и тестовый датафреймы из закодированных и отмасштабированных данных
X_train_ohe = pd.DataFrame(X_train_ohe, columns=encoder_col_names)
X_test_ohe = pd.DataFrame(X_test_ohe, columns=encoder_col_names)

X_train_scaled = pd.DataFrame(X_train_scaled, columns=num_col_names)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=num_col_names)

# объединяем все преобразованные данные в тренировочный и тестовый датафреймы
# сбрасываем индексы в датафрейме с числовыми индексами, чтобы избежать ошибок

X_train = pd.concat([X_train_ohe, X_train_scaled], axis=1)
X_test = pd.concat([X_test_ohe, X_test_scaled], axis=1)

# создаём объект для работы с моделью
clf = LogisticRegression(random_state=RANDOM_STATE) 

# обучаем модель на данных
clf = clf.fit(X_train, y_train) 

# получаем прогнозные метки классов
y_pred = clf.predict(X_test)
```

**Объясните, какая ошибка критичнее всего при покупке коров: первого или второго рода.**

Ошибки классификации включают в себя оценки двух видов: False Positive и False Negative (ошибки первого рода и второго рода) В задаче классификации молока ошибка первого рода — классифицировать корову со вкусным молоком как с не вкусным, а ошибка второго рода — определить корову с не вкусным молоком как со вкусным.

**Помните, что главное для заказчика — вкусное молоко. Сформулируйте ответ с учётом метрик recall и precision, выберите более важную из них.**

Precision оценивает точность, с которой модель присваивает объектам класс 1. Она реагирует на ошибки первого рода, однако не реагирует на ошибки второго.

Recall измеряет, смогла ли модель классификации присвоить класс 1 всем объектам этого класса. Её «слепое пятно» противоположно precision: она сосредоточена на ошибках второго рода и никак не учитывает ошибки первого.
 
В данном случае нам важнее метрика precision.
