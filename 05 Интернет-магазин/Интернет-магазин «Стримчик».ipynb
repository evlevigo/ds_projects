### Функция вывода информации о датафрейме

[К оглавлению](#Оглавление)

Разработаем функцию выводящую общую информацию о датафрейме а также функцию для форматирования таблиц

```python
def nice_t(data):
    th_props = [
      ('font-size', '11px'),
      ('text-align', 'center'),
      ('font-weight', 'bold'),
      ('color', 'black'),
      ('background-color', '#40E0D0')
      ]
    td_props = [
      ('font-size', '12px'),
      ('text-align', 'center')
      ]
    tr_props = [
      ('background-color', '#40E0D0'),
      ('text-align', 'center')  
        ]
    tr_nth_child_props = [
      ('background-color', '#CFECEC')  
        ]

    styles = [
      dict(selector="th", props=th_props),
      dict(selector="td", props=td_props),
      dict(selector="tr:nth-child(even)", props=tr_nth_child_props),
      dict(selector="tr:hover", props=tr_props)  
      ]
    display(data.style.set_table_styles(styles))
```

```python
def data_info (data):
    
    '''
    Функция выводит основную информацию о датафрейме
    в удобном формате
    '''
    print('\nОбщая информация о датафрейме .head()\n')
    nice_t(data.head(15))
    print('\nОбщая информация о датафрейме .info()\n')
    display(data.info())
    print('\nОбщая информация о датафрейме .describe()\n')
    print('\nКоличество пропусков в датафрейме\n', data.isna().sum())
    print('\nСоотношение пропусков в датафрейме, %\n', round(data.isna().sum() / len(data) * 100, 2))
    nice_t(data.describe())
    print('\nОбщая информация о датафрейме .hist()\n')
    data.hist(color='#000080',bins=50, figsize=(15, 10))
    plt.show()
```

Как видно, присутствуют числовые значения и есть значения `tbd`. Обратимся к словарю идиом от MIT: [расшифровка TBD](https://iso.mit.edu/idioms/tba-tbd/)

Ниже приведена выдержка из  источника:

 “TBA” and “TBD” mean “to be announced” and “to be determined,” respectively. These terms are used broadly in event planning; they indicate that although something is expected to happen, a particular aspect of that, such as place and time, remains to be confirmed.
 
Таким образом, аббревиатура по отношению к оценке пользователей может означать, что она на текущий момент не определена, но будет определена в будущем. В нашем случае это равнозначно тому, что данных нет.

Выясним, какое количество строк с такими значениями имеется

```python
df.query('user_score == "tbd"')['user_score'].count()
```

К сожалению, также как и с пустыми значениями, мы ничего не можем с ними поделать, поэтому, заменим их на -2 (чтобы было понимание, что это две разные категории - отсутствующие полностью значения и те, которые на текущий момент времени недоступны)

```python
df['user_score'].where(df['user_score'] != "tbd", -2, inplace=True)
```

```python
df['user_score'].unique()
```

Разработаем функцию для построения графика

```python
def top_plot (data, column_1, column_2, column_3, top_number, title, x_label, y_label):
    
    #функция выводит график значений с разбивкой по параметрам
    #column_1 колонка категории 1
    #column_2 колонка категории 2
    #column_1 и column_2 группируется
    #column_3 колонка
    
    count = 0
    summ = 0
    sns.set(rc = {'figure.figsize':(15,7)})
    #в переменную записывается аггрегированные данные с группировкой по column_1 и column_2
    analyze_data = data.pivot_table(index=[column_1,column_2], values=column_3, aggfunc = 'sum').reset_index()
    #в переменную записывается список для реализации цикла
    list_data = data.groupby(column_1).agg({column_3: 'sum'}). \
                sort_values(by=column_3, ascending=False).head(top_number).index.to_list()
    #Цикл для построения графиков
    for list_data in list_data:
        sns.lineplot(data=analyze_data[analyze_data[column_1] == list_data], 
                     x=column_2, 
                     y=column_3,
                     label=list_data,
                     marker='o')
        calc = analyze_data[analyze_data[column_1] == list_data][column_2].max() - \
               analyze_data[analyze_data[column_1] == list_data][column_2].min()
        print('Срок существования платформы',list_data, 'в выборке:', calc, 'лет')
        summ = calc + summ
        count += 1
    print ('Среднее время жизни: ', round(summ / count,2))    
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()    
```

Значение корреляции Пирсона характеризует уровень связи между переменными:

• 0,50 – 0.74 высокая положительная
• 0,25 – 0.49 средняя положительная
• 0,00 – 0.24 слабая положительная

• -0,50 – -0.74 высокая отрицательная

Коэффициент корреляции r - Пирсона оценивает только линейную связь

Подготовим переменные для анализа данных, исключая пропущенные значения

```python
critic_score_analyze = df_actual.query('critic_score >= 0')
user_score_analyze = df_actual.query('user_score >= 0')
```

Разработаем две функции, в процессе решим, какая из них больше подходит

```python
def corr_func(data, column_1, column_2, x_label, y_label, title):
    
    #Функция расчитывает корреляцию между двумя столбцами данных
    #и выводит диаграмму рассеяния
    
    #построение диаграммы рассеяния
    data.plot(
        x=column_1,
        y=column_2,
        kind='scatter',
        grid=True, 
        alpha=0.6,
        color='#003EFF')
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()
    
    #расчет коэффициента корреляции и вывод информации о нём
    print('Коэффициент корреляции')
    corr_coef = data[column_1].corr(data[column_2])
    print(corr_coef)
    if 0.75 <= corr_coef <= 1:
        print('Связь очень высокая положительная')
    if 0.50 <= corr_coef <= 0.74:
        print('Связь высокая положительная')        
    if 0.25 <= corr_coef <= 0.49:
        print('Связь средняя положительная')   
    if 0.00 <= corr_coef <= 0.24:
        print('Связь слабая положительная')   
    if 0.00 >= corr_coef >= -0.24:
        print('Связь слабая отрицательная')   
    if -0.25 >= corr_coef >= -0.49:
        print('Связь средняя отрицательная')   
    if -0.50 >= corr_coef >= -0.74:
        print('Связь высокая отрицательная')   
    if -0.75 >= corr_coef >= -1.00:
        print('Связь очень высокая отрицательная')           
```

Разработаем функцию для вывода информации

```python
def region_sales_analyze(data, column_1, column_2, x_label, y_label, title):
    
    '''
    Функция выводит сводную таблицу, диаграмму столбчатую и круговую 
    для исследования показателей по категориям 
    '''
    #создадим сводную таблицу
    region_sales = data.pivot_table(index=column_1, values=column_2, aggfunc=['sum'])
    region_sales.columns = ['Сумма']
    region_sales = region_sales.sort_values(by='Сумма', ascending = False).head(5)
    nice_t(region_sales)  
    
    #выводим столбчатую диаграмму
    region_sales.plot(kind='bar',figsize = (15,7), color='#00827F')
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show() 
    #выводим диаграмму круговую
    plt.pie(region_sales['Сумма'],
            labels = region_sales.index,
            colors=sns.color_palette('Spectral'),
            radius=2,
            autopct='%.1f%%',
            textprops={'size': 'medium'},) 
    plt.title(title, pad=100)
    plt.show()
```

## Шаг 6. Проверка гипотез

[К оглавлению](#Оглавление)

###  Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;

[К оглавлению](#Оглавление)

Задайте самостоятельно пороговое значение alpha.
Поясните:
- Как вы сформулировали нулевую и альтернативную гипотезы;
- Какой критерий применили для проверки гипотез и почему.

В Python есть специальный метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам: `scipy.stats.ttest_ind(array1, array2, equal_var, alternative)`

Чтобы понять, можно ли применять t-тест, нужно проверить три условия:

Генеральные совокупности не зависят друг от друга. Если вы рассматриваете одну генеральную совокупность до и после какого-либо изменения, нужно использовать другой тест. Речь о нём пойдёт в следующем уроке.

Выборочные средние нормально распределены. Это условие не должно быть препятствием благодаря ЦПТ. Если размер выборки составляет хотя бы несколько десятков значений, то выборочные средние, которые можно получить из одной и той же генеральной совокупности, будут распределены близко к нормальному вокруг истинного среднего этой совокупности. Это так, даже если сама генеральная совокупность не распределена нормально.

Дисперсии рассматриваемых генеральных совокупностей равны. С одной стороны, вы никогда точно не знаете, равны ли дисперсии рассматриваемых генеральных совокупностей. С другой — по умолчанию параметр equal_var принимает значение True: дисперсии считаются равными и тест объединяет обе выборки в одну, чтобы эту дисперсию поточнее оценить.При этом если выборки достаточно велики (30 и больше значений) и равны по размеру между собой, то такой подход оправдан. Симуляции, проведённые учёными, показывают, что даже если дисперсии на самом деле не равны, то тест редко ошибается.

Подготовим данные и посмотрим на их распределение

```python
pc_user_rating = df_actual.query('(platform == "PC") and (user_score > 0)')

xone_user_rating = df_actual.query('(platform == "XOne") and (user_score > 0)')
```

```python
sns.set(rc = {'figure.figsize':(15,7)})
sns.histplot(data=pc_user_rating, x='user_score', color="blue", alpha=0.95, label="Платформа PC", bins=50)
sns.histplot(data=xone_user_rating, x='user_score', color="skyblue", alpha=0.8, label="Платформа XOne", bins=50)
plt.legend() 
plt.title('Распределение оценок пользователя в зависимости от платформы')
plt.xlabel('Средняя оценка пользователей')
plt.ylabel('Количество')
plt.show()
```

Распределение нормальное, скошено влево (отрицательно ассиметричное). Данные независимые. Как видно, количество данных не сильно отличается. 

Сформулируем гипотезы:

- H<sub>0</sub>: `пользовательский рейтинг PC` = `пользовательский рейтинг XOne`
- H<sub>1</sub>: `пользовательский рейтинг PC` &#8800; `пользовательский рейтинг XOne`

```python
alpha = 0.05 # уровень статистической значимости
# если p-value окажется меньше него, отвергнем гипотезу

results = st.ttest_ind(pc_user_rating['user_score'], xone_user_rating['user_score'])

print('p-значение:', round(results.pvalue,5))

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

```python
print('\nОбщая информация по платформе PC\n')
print(pc_user_rating['user_score'].describe())
print('\nОбщая информация по платформе XOne\n')
print(xone_user_rating['user_score'].describe())
```
